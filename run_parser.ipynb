```python
import parser
```

```python
def parse(sentence, language="English", p=0.1, LEX_k=20, 
          project_rounds=20, verbose=True, debug=False, readout_method=parser.ReadoutMethod.FIBER_READOUT):
    if language == "English":
        b = parser.EnglishParserBrain(p, LEX_k=LEX_k, verbose=verbose)
        lexeme_dict = parser.LEXEME_DICT
        all_areas = parser.AREAS
        explicit_areas = parser.EXPLICIT_AREAS
        readout_rules = parser.ENGLISH_READOUT_RULES

    parser.parseHelper(b, sentence, p, LEX_k, project_rounds, verbose, debug,
                       lexeme_dict, all_areas, explicit_areas, readout_method, readout_rules)
```

```python
parse("the cats chase mice")
```
